```
import tensorflow_hub as hub

module_handler = "https://tfhub.dev/google/faster_rcnn/openimages_v4/inception_resnet_v2/1"
detector = hub.load(module_handler).signatures['default']
```

```
import tensorflow as tf

x_1 = tf.io.read_file("/content/IMG20200803154929.jpg")
x_1 = tf.image.decode_jpeg(x_1,channels=3)

img = tf.image.convert_image_dtype(x_1,tf.float32)[tf.newaxis, ...]
x = detector(img)
```

```
result = {
    key:value.numpy() for key,value in x.items()
}
print(result)
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from PIL import ImageColor
from PIL import ImageDraw
from PIL import ImageFont

def draw_boxes(img,boxes,classes,scores,max_boxes=50,min_score=0.5):
  print("hai_1")
  for i in range(max_boxes):
    if scores[i] >= min_score:
      y_min, x_min, y_max, x_max = tuple(boxes[i])
      img_1 = cv2.imread("/content/IMG20200803154929.jpg")
      print(img_1.shape)
      x_pil = Image.fromarray(img).convert("RGB")

      #x_pil = np.array(img)
      str_1 = classes[i].decode("ascii")+':'+str(int(scores[i]*10))
      draw_bounding_boxes(x_pil,y_min,x_min,y_max,x_max,str_1)
      np.copyto(img,np.array(x_pil))
  return img
      
def draw_bounding_boxes(img,y_min,x_min,y_max,x_max,str_1):
  draw = ImageDraw.Draw(img)
  im_width, im_height = img.size
  print(img.size)
  (left,right,top,bottom) = (x_min*im_width,x_max*im_width,
                             y_min*im_height,y_max*im_height)
  draw.line(
      [(left,top),(left,bottom),(right,bottom),(right,top),(left,top)],width=10
  )

  #draw.rectangle([(left,top),(left+500,top-250)],outline='red')
  font = font = ImageFont.truetype("/usr/share/fonts/truetype/liberation/LiberationSansNarrow-Regular.ttf",
                              100)
  draw.text((left,top),str_1,fill='red',font = font)
  plt.figure(figsize=(20,10))
  plt.imshow(img)

print("hai")
image_with_boxes = draw_boxes(x_1.numpy(),result["detection_boxes"],result["detection_class_entities"],result["detection_scores"])
```
